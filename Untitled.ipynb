```python
import pandas as pd
import numpy as np

# Creating sample real estate data
np.random.seed(42)
n = 500

df = pd.DataFrame({
    'city': np.random.choice(['Mumbai', 'Pune', 'Delhi', 'Bangalore', 'Chennai'], n),
    'price': np.random.randint(2000000, 15000000, n),
    'bedrooms': np.random.choice([1, 2, 3, 4, 5], n),
    'bathrooms': np.random.choice([1, 2, 3], n),
    'sqft_living': np.random.randint(500, 4000, n),
    'yr_built': np.random.randint(1990, 2023, n)
})

print(df.shape)
df.head()
```

```python
# Clean data
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)

print("Shape after cleaning:", df.shape)
print("\nData types:")
print(df.dtypes)
```

```python
# Average price per city
avg_price = df.groupby('city')['price'].mean().sort_values(ascending=False)

print("Average Property Price by City:")
print(avg_price)
```

```python
# Bar chart
avg_price.plot(kind='bar', figsize=(10, 5), color='steelblue')

import matplotlib.pyplot as plt
plt.title('Average Property Price by City')
plt.ylabel('Price (INR)')
plt.xlabel('City')
plt.tight_layout()
plt.savefig('price_by_city.png')
plt.show()

print("Chart saved as price_by_city.png")
```
